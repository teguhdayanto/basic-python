### Dasar-dasar Matplotlib


#### Apa itu matplotlib?

Matplotlib adalah *library* plotting grafik tingkat rendah dengan python yang berfungsi sebagai utilitas visualisasi.

Matplotlib dibuat oleh John D. Hunter.

Matplotlib bersifat open source dan dapat kita gunakan secara bebas.

Matplotlib sebagian besar ditulis dengan python, beberapa segmen ditulis dalam C, Objective-C dan Javascript untuk kompatibilitas Platform.


#### Di mana Basis Kode Matplotlib?

Kode sumber untuk Matplotlib terletak di repositori github: https://github.com/matplotlib/matplotlib

image.png

Sekarang paket Pyplot dapat disebut sebagai `plt`.

### Plotting


#### Plotting titik x dan y

Fungsi `plot()` digunakan untuk menggambar titik (marker) pada diagram.

Secara default, fungsi `plot()` menarik garis dari titik ke titik.

Fungsi mengambil parameter untuk menentukan titik dalam diagram.

Parameter 1 adalah larik yang berisi titik-titik pada **sumbu x**.

Parameter 2 adalah larik yang berisi titik-titik pada **sumbu y**.

Jika kita perlu memplot garis dari (1, 3) ke (8, 10), kita harus meneruskan dua larik [1, 8] dan [3, 10] ke fungsi plot.


```python
# Gambar garis dalam diagram dari posisi (1, 3) ke posisi (8, 10)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

x-axis (sumbu-x) adalah sumbu horisontal.

y-axis (sumbu-y) adalah sumbu vertikal.


#### Plot tanpa garis

Untuk memplot penanda saja, kita dapat menggunakan pintasan parameter notasi string (string notation parameter) 'o', yang berarti 'rings'.


```python
# Gambarlah dua titik pada diagram, satu di posisi (1, 3) dan satu di posisi (8, 10)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

#### Titik-titik ganda

Kita dapat memplot titik sebanyak yang kita suka, pastikan kita memiliki jumlah titik yang sama di kedua sumbu.


```python
# Buat garis dalam diagram dari posisi (1, 3) ke (2, 8) lalu ke (6, 1) dan terakhir ke posisi (8, 10)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

#### Default Titik-titik X

Jika kita tidak menentukan titik pada sumbu x, mereka akan mendapatkan nilai default 0, 1, 2, 3 (dst., bergantung pada panjang titik y.

Jadi, jika kita mengambil contoh yang sama seperti di atas, dan menghilangkan titik-x, diagram akan terlihat seperti ini.


```python
# Plot tanpa titik-titik - x

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

Titik-titik X dalam contoh diatas adalah [0, 1, 2, 3, 4, 5].


### Marker (Penanda)



#### Markers

Kita dapat menggunakan argumen kata kunci `marker` untuk menekankan setiap titik dengan penanda tertentu.


```python
# Tandai setiap titik dengan circle (lingkaran)

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
# Tandai setiap titik dengan circle (lingkaran)

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '*')
plt.show()
```

#### Referensi penanda (marker)

Kita bebas dapat menggunakan marker (penanda) berikut.


image.png

#### Format string `fmt`

Kita juga dapat menggunakan parameter notasi string pintasan (*shortcut string notation*) untuk menentukan penanda.

Parameter ini juga disebut `fmt`, dan ditulis dengan sintaks berikut:

`marker|line|color`


```python
# Tandai setiap titik dengan lingkaran

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

Nilai marker bisa apa saja dari Referensi Marker di atas.

Nilai baris dapat berupa salah satu dari berikut ini:


image.png

**Catatan:** Jika kita tidak memasukkan nilai garis dalam parameter fmt, tidak ada garis yang akan diplot.


Warna dapat diperoleh dari salah satu berikut.

#### Marker Size (Ukuran Penanda)

Kita dapat menggunakan argumen kata kunci `markersize` atau versi yang lebih pendek, `ms` untuk menyetel ukuran marker.


```python
# Atur ukuran marker (penanda) ke 20

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

#### Marker Color (Warna Penanda)

Kita dapat menggunakan argumen kata kunci `markeredgecolor` atau `mec` yang lebih pendek untuk menyetel warna tepi penanda.


```python
# Atur warna EDGE (tepi) ke merah

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

Kita dapat menggunakan argumen kata kunci `markerfacecolor` atau `mfc` yang lebih pendek, untuk menyetel warna di dalam tepi penanda.


```python
# Atur warna FACE (tengah) ke merah

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

Gunakan argumen mec dan mfc untuk mewarnai seluruh marker (penanda).


```python
# Atur warna dari kedua edge (tepi) dan face (tengah) ke merah

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
```

Kita juga dapat menggunakan nilai warna Hexadecimal.


```python
# Tandai setiap titik dengan sebuah warna hijau cantik

plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
```

Atau bisa diambil dari 140 nama-nama warna yang support.


```python
# Tandai setiap titik dengan warna "hotpink"

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'hotpink', mfc = 'hotpink')
```

### Line (Garis)


#### Linestyle

Kita bisa menggunakan argumen kata kunci `linestyle`, atau lebih pendek `ls`, untuk mengubah gaya garis yang diplot.


```python
# Gunakan sebuah dotted line

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
# Gunakan sebuah dashed line

plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

#### Shorter syntax (sintaks yang lebih pendek)

Line style dapat ditulis dalam sebuah sintaks yang lebih pendek.

`linestyle` dapat ditulis sebagai `ls`.

`dotted` dapat ditulis sebagai `:`.

`dashed` dapat ditulis sebagai `--`.


```python
import sys
import matplotlib
matplotlib.use('Agg')

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls = ':')
plt.show()
```

Kita dapat memilih linestyles dari daftar pilihan berikut.


image.png

#### Line color (warna garis)

Kita dapat menggunakan argumen kata kunci `color` atau lebih pendek `c` untuk mengatur warna garis.


```python
# Atur warna garis ke merah

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'r')
plt.show()
```

Kita dapat juga menggunakan nilai warna Hexadecimal.


```python
# Plot dengan sebuah garis hijau cantik

plt.plot(ypoints, c = '#4CAF50')
```

Atau kita bisa mengambil dari satu diantara 140 nama-nama warna yang support.


```python
# Plot dengan warna "hotpink"

plt.plot(ypoints, c = 'hotpink')
```

#### Line Width (Lebar Garis)

Kita dapat menggunakan argumen kata kunci `linewidth` atau lw yang lebih pendek untuk mengubah lebar garis.

Nilainya adalah angka *floating*, dalam ukuran points.


```python
# Plot dengan garis selebar 20.5pt

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

#### Multiple lines (Garis berganda)

Kita dapat memplot baris sebanyak yang kita suka hanya dengan menambahkan lebih banyak fungsi `plt.plot()`.


```python
# Gambar 2 garis dengan menggunakan fungsi plt.plot() untuk setiap garis

import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

Kita juga dapat memplot banyak garis dengan menambahkan titik sumbu x dan y untuk setiap baris dalam fungsi `plt.plot()` yang sama.

(Dalam contoh di atas, kami hanya menentukan titik pada sumbu y, artinya titik pada sumbu x mendapatkan nilai default (0, 1, 2, 3).)

Nilai x dan y datang berpasangan.


```python
# Gambar 2 garis dengan menentukan nilai x- and y-point untuk kedua garis

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

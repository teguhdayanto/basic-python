# A. Join data csv menggunakan pandas


## A.1. Import library pandas


```python
import pandas as pd
```

2. Upload file csv

* sample-listing-properti-1.csv
* sample-listing-properti-2.csv


```python
df1 = pd.read_csv('sample-listing-properti-1.csv')
df2 = pd.read_csv('sample-listing-properti-2.csv')
```

```python
df1
```

```python
df2
```

## A.2. Inner Join

Inner Join merupakan perintah untuk menampilkan semua data yang mempunyai nilai sama dari dua table ataupun lebih.


```python
inner_join = pd.merge(df1, 
                      df2, 
                      on ='Agent', 
                      how ='inner')

# Tampilkan tabel
inner_join
```

```python
# Simpan ke file csv dengan nama 'inner_join.csv'
inner_join.to_csv(r'inner_join.csv', index=False)
```

## A.3. Outer Join

Pada Outer Join, data pada kedua tabel akan ditampilkan semua.


```python
outer_join = pd.merge(df1, 
                      df2, 
                      on ='Agent', 
                      how ='outer')

# Tampilkan tabel
outer_join
```

```python
# Simpan ke file csv dengan nama 'outer_join.csv'
outer_join.to_csv(r'outer_join.csv', index=False)
```

## A.4. Left Join

Left Join merupakan cara menghubungkan tabel dan menampilkan semua data (kiri) pada tabel yang tidak berhubungan, sedangkan data yang kosong akan bernilai NULL.

Dalam hal ini, df1 sebagai referensi.


```python
left_join = pd.merge(df1, 
                     df2, 
                     on ='Agent', 
                     how ='left')

# Tampilkan tabel
left_join
```

```python
# Simpan ke file csv dengan nama 'left_join.csv'
left_join.to_csv(r'left_join.csv', index=False)
```

## A.5. Right Join

Right Join, fungsi ini hampir sama dengan fungsi Left Join ataupun kebalikannya, dimana Right Join akan menghubungkan tabel dan menampilkan semua data (kanan) pada tabel yang tidak berhubungan, dan data yang kosong akan bernilai NULL.


```python
right_join = pd.merge(df1, 
                      df2, 
                      on ='Agent',
                      how ='right')

# Tampilkan tabel
right_join
```

```python
# Simpan ke file csv dengan nama 'right_join.csv'
right_join.to_csv(r'right_join.csv', index=False)
```

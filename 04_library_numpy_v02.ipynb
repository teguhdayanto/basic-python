##### Periksa Jumlah Dimensi Array

NumPy Arrays menyediakan atribut `ndim` yang mengembalikan bilangan bulat yang memberi tahu kita berapa banyak dimensi yang dimiliki array.


```python
# Periksa berapa jumlah dimensi yang dimiliki array berikut

import numpy as np

a = np.array(42)
b = np.array([1, 2, 3, 4, 5])
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)
```

##### Array Dimensi Tinggi

Array dapat memiliki sejumlah dimensi.

Saat array dibuat, kita dapat menentukan jumlah dimensi dengan menggunakan argumen `ndmin`.


```python
# Buat sebuah array 5-D dan verifikasi bahwa benar memiliki 5 dimensi

import numpy as np

arr = np.array([1, 2, 3, 4], ndmin=5)

print(arr)
print('number of dimensions :', arr.ndim)
```

Pada array ini dimensi terdalam (dimensi ke-5) memiliki 4 elemen, dim ke-4 memiliki 1 elemen yaitu vektor, dim ke-3 memiliki 1 elemen yaitu matriks dengan vektor, dim ke-2 memiliki 1 elemen yaitu array 3D dan Dim ke-1 memiliki 1 elemen yaitu array 4D.

### Slicing (pemotongan) array



##### Pemotongan array

Memotong (mengiris) dengan python berarti mengambil elemen dari satu indeks yang diberikan ke indeks lain yang diberikan.

Kita memberikan slice alih-alih indeks seperti ini: `[start:end]`.

Kita juga bisa menentukan langkahnya, seperti ini: `[start:end:step]`.

Jika kita tidak lolos mulai dianggap 0

Jika kita tidak lolos end dianggap panjang array dalam dimensi itu

Jika kita tidak lolos langkah dianggap 1


```python
# Iris/potong elemen dari indeks 1 ke indeks 5 dari array berikut

import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6, 7])

print(arr[1:5])
```

**Catatan:** Hasilnya mencakup indeks awal, tetapi tidak termasuk indeks akhir.


```python
# Iris/potong elemen dari indeks 4 ke akhir dari array

import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6, 7])

print(arr[4:])
```

```python
# Iris/potong elemen dari awal ke indeks 4 (indeks 4 tidak termasuk)

import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6, 7])

print(arr[:4])
```

##### Pemotongan negatif

Gunakan operator minus untuk merujuk ke indeks dari akhir.


```python
# Iris/potong dari indeks 3 dari akhir ke indeks 1 dari akhir

import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6, 7])

print(arr[-3:-1])
```

```python
# Iris/potong dari indeks 3 dari akhir ke indeks terakhir (dimasukkan)

import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6, 7])

print(arr[-3:])
```

##### STEP (Langkah)

Gunakan nilai `step` untuk menentukan berapa langkah dari suatu pemotongan/pengirisan.


```python
# Kembalikan setiap elemen dari indeks 1 ke indeks 5 dengan step 2

import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6, 7])

print(arr[1:5:2])
```

```python
# Kembalikan setiap elemen dari seluruh array

import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6, 7])

print(arr[::2])
```

### Tipe Data NumPy


#### Tipe Data NumPy di Python

Secara default Python memiliki tipe data ini:

`strings` - digunakan untuk mewakili data teks, teks diberikan di bawah tanda kutip. misalnya "ABCD"

`integer` - digunakan untuk mewakili bilangan bulat. misalnya -1, -2, -3

`float` - digunakan untuk mewakili bilangan real. misalnya 1.2, 42.42

`boolean` - digunakan untuk mewakili True atau False.

`complex` - digunakan untuk mewakili bilangan kompleks. misalnya 1.0 + 2.0j, 1.5 + 2.5j

NumPy memiliki beberapa tipe data tambahan, dan mengacu pada tipe data dengan satu karakter, seperti `i` untuk integer, `u` untuk unsigned integer dll.

#### Membuat array dengan tipe data yang didefinisikan

Kita menggunakan fungsi `array()` untuk membuat array, fungsi ini dapat menggunakan argumen opsional: `dtype` yang memungkinkan kami untuk menentukan tipe data yang diharapkan dari elemen array.

#### Bagaimana jika Nilai Tidak Dapat Dikonversi?

Jika suatu tipe diberikan di mana elemennya tidak dapat dimasukkan maka NumPy akan memunculkan ValueError.

ValueError: Dalam Python ValueError dimunculkan ketika jenis argumen yang diteruskan ke suatu fungsi tidak terduga/salah.


```python
# String bukan bilangan bulat seperti 'a' tidak dapat diubah menjadi bilangan bulat
# dan akan menimbulkan kesalahan (error)

import numpy as np

arr = np.array(['a', '2', '3'], dtype='i')
```

#### Mengonversi Tipe Data pada Array yang Ada

Cara terbaik untuk mengubah tipe data dari array yang ada, adalah membuat salinan dari array dengan metode `astype()`.

Fungsi `astype()` membuat salinan array, dan memungkinkan kita menentukan tipe data sebagai parameter.

Tipe data dapat ditentukan menggunakan string, seperti 'f' untuk float, 'i' untuk integer dll atau kita dapat menggunakan tipe data secara langsung seperti float untuk `float` dan `int` untuk integer.


```python
# Ubah tipe data dari float ke integer dengan menggunakan 'i' sebagai nilai parameter

import numpy as np

arr = np.array([1.1, 2.1, 3.1])

newarr = arr.astype('i')

print(newarr)
print(newarr.dtype)
```

```python
# Ubah tipe data dari float ke integer dengan menggunakan 'int' sebagai nilai parameter

import numpy as np

arr = np.array([1.1, 2.1, 3.1])

newarr = arr.astype(int)

print(newarr)
print(newarr.dtype)
```

```python
# Ubah tipe data dari integer ke boolean

import numpy as np

arr = np.array([1, 0, 3, -2, 0])

newarr = arr.astype(bool)

print(newarr)
print(newarr.dtype)
```

### Pengenalan Random


#### Apa itu sebuah angka random?

Angka acak TIDAK berarti angka yang berbeda setiap saat. Acak berarti sesuatu yang tidak dapat diprediksi secara logis.


#### Pseudo random dan true random

Komputer bekerja pada program, dan program adalah kumpulan instruksi yang pasti. Jadi ini berarti harus ada beberapa algoritma untuk menghasilkan angka acak juga.

Jika ada program untuk membangkitkan bilangan acak maka dapat diprediksi, sehingga tidak benar-benar acak.

Angka acak yang dihasilkan melalui algoritma pembangkitan disebut acak semu.

Bisakah kita membuat angka yang benar-benar acak?

Ya. Untuk menghasilkan angka yang benar-benar acak di komputer kita, kita perlu mendapatkan data acak dari beberapa sumber luar. Sumber luar ini umumnya adalah penekanan tombol, gerakan mouse, data di jaringan, dll.

Kami tidak memerlukan angka yang benar-benar acak, kecuali jika terkait dengan keamanan (mis. *encryption keys*) atau dasar penerapannya adalah keacakan (mis. Roda roulette digital).

Dalam tutorial ini kita akan menggunakan bilangan acak semu.


#### Menghasilkan Nomor Acak (Random Number)

NumPy menawarkan modul random/acak untuk bekerja dengan angka acak.


```python
# Hasilkan sebuah integer random dari 0 sampai 100

from numpy import random

x = random.randint(100)

print(x)
```

#### Menghasilkan Float Acak (Random Float)

Metode `rand()` modul acak mengembalikan float acak antara 0 dan 1.


```python
# HAsilkan sebuah random float dari 0 sampai 1

from numpy import random

x = random.rand()

print(x)
```

#### Menghasilkan Array Acak (Random Array)

Di NumPy kita bekerja dengan array, dan kita dapat menggunakan dua metode dari contoh di atas untuk membuat array acak.

**Bilangan bulat**

Metode `randint()` mengambil parameter `size` tempat kita dapat menentukan bentuk array.


```python
# Hasilkan sebuah array 1-D yang mengandung 5 angka integer random dari 0 sampai 100

from numpy import random

x=random.randint(100, size=(5))

print(x)
```

#### Menghasilkan Nomor Acak Dari Array

Metode `choice()` memungkinkan kita menghasilkan nilai acak berdasarkan array nilai.

Metode `choice()` mengambil array sebagai parameter dan secara acak mengembalikan salah satu nilai.


```python
# Kembalikan satu dari nilai dalam sebuah array

from numpy import random

x = random.choice([3, 5, 7, 9])

print(x)
```

Metode `choice()` juga memungkinkan kita mengembalikan array nilai.

Tambahkan parameter `size` untuk menentukan bentuk larik (array).

### Distribusi Data



**Apa itu Distribusi Data?**

Distribusi Data adalah daftar semua kemungkinan nilai, dan seberapa sering setiap nilai muncul.

Daftar semacam itu penting saat bekerja dengan statistik dan ilmu data.

Modul acak menawarkan metode yang mengembalikan distribusi data yang dihasilkan secara acak.

**Distribusi Acak**

Distribusi acak adalah sekumpulan angka acak yang mengikuti fungsi kepadatan probabilitas tertentu.

**Fungsi Densitas Probabilitas**: Fungsi yang menggambarkan probabilitas kontinu. yaitu probabilitas semua nilai dalam array.

Kita dapat menghasilkan angka acak berdasarkan probabilitas yang ditentukan menggunakan metode `choice()` dari modul `random`.

Metode `choice()` memungkinkan kita menentukan probabilitas untuk setiap nilai.

Probabilitas ditentukan oleh angka antara 0 dan 1, dimana 0 berarti nilai tidak akan pernah muncul dan 1 berarti nilai akan selalu muncul.


```python
# Hasilkan array 1-D yang berisi 100 nilai, di mana setiap nilai harus 3, 5, 7 atau 9.

# Probabilitas untuk nilai menjadi 3 diatur menjadi 0,1

# Probabilitas untuk nilai menjadi 5 diatur menjadi 0,3

# Probabilitas untuk nilai menjadi 7 diatur menjadi 0,6

# Probabilitas untuk nilai menjadi 9 diatur menjadi 0

from numpy import random

x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(100))

print(x)
```

Jumlah semua angka probabilitas harus 1.

Bahkan jika kita menjalankan contoh di atas 100 kali, nilai 9 tidak akan pernah muncul.

Kita dapat mengembalikan array dalam bentuk dan ukuran apa pun dengan menentukan bentuk dalam parameter ukuran.


```python
# Sama seperti contoh diatas. tetapi mengembalikan sebuah array 2-D dengan 3 baris, masing-masing mengandung 5 nilai

from numpy import random

x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5))

print(x)
```

### Permutasi Random


#### Permutasi random dari elemen

Permutasi mengacu pada susunan elemen. misalnya [3, 2, 1] adalah permutasi dari [1, 2, 3] dan sebaliknya.

Modul NumPy Random menyediakan dua metode untuk ini: `shuffle()` dan `permutation()`.


#### Mengacak (Suffling) Array

Shuffle berarti mengubah susunan elemen di tempat. yaitu dalam array itu sendiri.


```python
# Acak elemen array berikut secara acak

from numpy import random
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

random.shuffle(arr)

print(arr)
```

Metode `shuffle()` membuat perubahan pada array asli.


#### Membangkitkan Permutasi Array

### Distribusi Normal (Gaussian)


#### Distribusi Normal

Distribusi Normal adalah salah satu distribusi yang paling penting.

Itu juga disebut Distribusi Gaussian setelah matematikawan Jerman Carl Friedrich Gauss.

Ini cocok dengan distribusi probabilitas banyak peristiwa, misalnya: Skor IQ, Detak Jantung, dll.

Gunakan metode `random.normal()` untuk mendapatkan Distribusi Data Normal.

Distribusi Normal memiliki tiga parameter:

`loc` - (Berarti) di mana puncak bel ada.

`scale` - (Standar Deviasi) seberapa datar distribusi grafik seharusnya.

`size` - Bentuk array yang dikembalikan.


```python
# Bangkitkan sebuah distribusi normal random dengan ukuran 2x3

from numpy import random

x = random.normal(size=(2, 3))

print(x)
```

```python
# Bangkitkan sebuah distribusi normal random dengan ukuran 2x3
# dengan nilai rata-rata dan standar deviasi 2

from numpy import random

x = random.normal(loc=1, scale=2, size=(2, 3))

print(x)
```

#### Visualisasi Distribusi Normal


```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(size=1000), hist=False)

plt.show()
```

**Catatan:** Kurva Distribusi Normal juga dikenal sebagai Kurva Lonceng karena kurvanya berbentuk lonceng.

## **NumPy ufunc**


### Pengenalan


#### Apa itu ufunc?

ufuncs adalah singkatan dari "Universal Functions" dan merupakan fungsi NumPy yang beroperasi pada objek `ndarray`.


#### Mengapa menggunakan ufunc?

ufunc digunakan untuk mengimplementasikan vektorisasi di NumPy yang jauh lebih cepat daripada mengulang elemen.

Mereka juga menyediakan penyiaran dan metode tambahan seperti pengurangan, akumulasi, dll. yang sangat membantu perhitungan.

ufuncs juga mengambil argumen tambahan, seperti:

`where` - merupakan array boolean atau kondisi yang menentukan di mana operasi harus dilakukan.

`dtype` - mendefinisikan tipe pengembalian elemen.

`out` - merupakan array keluaran di mana nilai pengembalian harus disalin.


#### Apa itu vektorisasi?

Mengubah pernyataan iteratif menjadi operasi berbasis vektor disebut vektorisasi.

Ini lebih cepat karena CPU modern dioptimalkan untuk operasi semacam itu.

Tambahkan Elemen dari Dua Daftar
daftar 1: [1, 2, 3, 4]

daftar 2: [4, 5, 6, 7]

Salah satu cara melakukannya adalah dengan mengulangi kedua list dan kemudian menjumlahkan setiap elemen.


```python
# Tanpa ufunc, kita dapat menggunakan metode Python's built-in zip()

x = [1, 2, 3, 4]
y = [4, 5, 6, 7]
z = []

for i, j in zip(x, y):
  z.append(i + j)
print(z)
```

NumPy memiliki ufunc untuk ini, disebut `add(x, y)` yang akan menghasilkan hasil yang sama.


```python
# Dengan ufunc, kita dapat menggunakan fungsi add()

import numpy as np

x = [1, 2, 3, 4]
y = [4, 5, 6, 7]
z = np.add(x, y)

print(z)
```

### Membuat Fungsi Kita Sendiri


#### Bagaimana cara kita membuat fungsi sendiri?

Untuk membuat ufunc kita sendiri, kita harus mendefinisikan sebuah fungsi, seperti yang kita lakukan dengan fungsi normal di Python, lalu kita menambahkannya ke pustaka ufunc NumPy Anda dengan metode `frompyfunc()` .

Metode `frompyfunc()` mengambil argumen berikut:

`function` - nama fungsi.

`inputs` - jumlah argumen input (array).

`outputs` - jumlah array output.


```python
# Buat ufunc kita sendiri untuk penambahan.

import numpy as np

def myadd(x, y):
  return x+y

myadd = np.frompyfunc(myadd, 2, 1)

print(myadd([1, 2, 3, 4], [5, 6, 7, 8]))
```

#### Periksa jika sebuah Fungsi adalah sebuah ufunc

Periksa jenis fungsi untuk memeriksa apakah itu ufunc atau bukan.

Ufunc harus mengembalikan `<class 'numpy.ufunc'>`.


```python
# Periksa jika sebuah fungsi adalah sebuah ufunc

import numpy as np

print(type(np.add))
```

Jika ini bukan ufunc, ini akan mengembalikan tipe lain, seperti fungsi NumPy bawaan ini untuk menggabungkan dua atau lebih array.


```python
# Periksa tipe dari fungsi yang lain yaitu concatenate()

import numpy as np

print(type(np.concatenate))
```

Jika fungsinya tidak dikenali sama sekali, itu akan mengembalikan error.


```python
# Periksa tipe dari sesuatu yang gak ada. Ini akan menghasilkan sebuah error

import numpy as np

print(type(np.blahblah))
```

Untuk menguji apakah fungsi adalah ufunc dalam pernyataan if, gunakan nilai `numpy.ufunc` (atau `np.ufunc` jika Anda menggunakan np sebagai alias untuk numpy).


```python
# Gunakan sebuah pernyataan if untuk mengecek jika fungsi adalah sebuah ufunc atau bukan

import numpy as np

if type(np.add) == np.ufunc:
  print('add is ufunc')
else:
  print('add is not ufunc')
```

### Aritmetika Sederhana


#### Aritmetika Sederhana

Kita dapat menggunakan operator aritmatika `+ - * /` langsung di antara array NumPy, tetapi bagian ini membahas ekstensi yang sama di mana kita memiliki fungsi yang dapat mengambil objek seperti array, mis: list, tuple, dll dan melakukan aritmatika secara kondisional.

**Arithmetic Conditionally:** artinya kita dapat menentukan kondisi dimana operasi aritmatika harus terjadi.

Semua fungsi aritmatika yang dibahas mengambil parameter `where` di mana kita dapat menentukan kondisi tersebut.


#### Penambahan

Fungsi `add()` menjumlahkan konten dari dua larik, dan mengembalikan hasilnya dalam larik baru.


```python
# Tambahkan nilai dalam arr1 kedalam nilai didalam arr2:

import numpy as np

arr1 = np.array([10, 11, 12, 13, 14, 15])
arr2 = np.array([20, 21, 22, 23, 24, 25])

newarr = np.add(arr1, arr2)

print(newarr)
```

Contoh di atas akan mengembalikan [30 32 34 36 38 40] yang merupakan jumlah dari 10+20, 11+21, 12+22 dll.


#### Pengurangan

Fungsi `subtract()` mengurangi nilai dari satu larik (array) dengan nilai dari larik (array) lain, dan mengembalikan hasilnya dalam larik (array) baru.


```python
# Kurangkan nilai di arr1 dengan nilai di arr2

import numpy as np

arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([20, 21, 22, 23, 24, 25])

newarr = np.subtract(arr1, arr2)

print(newarr)
```

Contoh di atas akan mengembalikan [-10 -1 8 17 26 35] yang merupakan hasil dari 10-20, 20-21, 30-22 dst.


#### Perkalian

Fungsi `multiply()` mengalikan nilai dari satu larik (array) dengan nilai dari larik (array) lain, dan mengembalikan hasilnya dalam larik (array) baru.


```python
# Kalikan nilai di arr1 dengan nilai di arr2

import numpy as np

arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([20, 21, 22, 23, 24, 25])

newarr = np.multiply(arr1, arr2)

print(newarr)
```

Contoh di atas akan mengembalikan [200 420 660 920 1200 1500] yang merupakan hasil dari 10*20, 20*21, 30*22 dll.


#### Pembagian

Fungsi `divide()` membagi nilai dari satu larik (array) dengan nilai dari larik (array) lain, dan mengembalikan hasilnya dalam larik (array) baru.


```python
# Bagi nilai di arr1 dengan nilai di arr2

import numpy as np

arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 5, 10, 8, 2, 33])

newarr = np.divide(arr1, arr2)

print(newarr)
```

Contoh di atas akan mengembalikan [3.33333333 4.3.5.25.1.81818182] yang merupakan hasil dari 10/3, 20/5, 30/10 dst.


#### Pangkat (Power)

Fungsi `power()` menaikkan nilai dari larik (array) pertama ke pangkat nilai larik (array) kedua, dan mengembalikan hasilnya dalam larik (array) baru.


```python
# Naikkan nilai di arr1 dengan pangkat dari nilai di arr2

import numpy as np

arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 5, 6, 8, 2, 33])

newarr = np.power(arr1, arr2)

print(newarr)
```

Contoh di atas akan mengembalikan [1000 3200000 729000000 6553600000000 2500 0] yang merupakan hasil dari 10*10*10, 20*20*20*20*20, 30*30*30*30*30*30 dll.


#### Sisa

Baik fungsi `mod()` maupun `remainder()` mengembalikan sisa nilai dalam larik (array) pertama yang sesuai dengan nilai dalam larik (array) kedua, dan mengembalikan hasilnya dalam larik (array) baru.


```python
# Kembalikan sisa

import numpy as np

arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 7, 9, 8, 2, 33])

newarr = np.mod(arr1, arr2)

print(newarr)
```

Contoh di atas akan mengembalikan [1 6 3 0 0 27] yang merupakan sisa ketika kita membagi 10 dengan 3 (10%3), 20 dengan 7 (20%7) 30 dengan 9 (30%9) dll.


Kita akan mendapatkan hasil yang sama saat menggunakan fungsi `remainder()`.


```python
# Kembalikan sisa

import numpy as np

arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 7, 9, 8, 2, 33])

newarr = np.remainder(arr1, arr2)

print(newarr)
```

#### Hasil bagi (quotient) dan Mod

Fungsi `divmod()` mengembalikan hasil bagi dan mod. Nilai yang dikembalikan adalah dua larik (array), larik (array) pertama berisi hasil bagi dan larik (array) kedua berisi mod (sisa dari hasil baginya).


```python
# Kembalikan quotient (hasil bagi) dan mod

import numpy as np

arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 7, 9, 8, 2, 33])

newarr = np.divmod(arr1, arr2)

print(newarr)
```

Contoh di atas akan mengembalikan:
(larik([3, 2, 3, 5, 25, 1]), larik([1, 6, 3, 0, 0, 27]))

Larik (array) pertama mewakili hasil bagi, (nilai bilangan bulat saat kita membagi 10 dengan 3, 20 dengan 7, 30 dengan 9, dll.

Larik (array) kedua mewakili sisa dari divisi (hasil pembagian) yang sama.
